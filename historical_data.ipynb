```python
import pandas as pd
import yfinance as yf
import datetime
import time
import requests
import io
```

```python
start = datetime.datetime(2020,2,1)
end = datetime.datetime(2023,10,10)
```

```python
url="https://pkgstore.datahub.io/core/nasdaq-listings/nasdaq-listed_csv/data/7665719fb51081ba0bd834fde71ce822/nasdaq-listed_csv.csv"
s = requests.get(url).content
companies = pd.read_csv(io.StringIO(s.decode('utf-8')))
```

```python
Symbols = companies['Symbol'].tolist()
```

```python
stock_final = pd.DataFrame()
for i in Symbols:  
# print the symbol which is being downloaded
#     print( str(Symbols.index(i)) + str(' : ') + i, sep=',', end=',', flush=True)  
    
    try:
        # download the stock price 
        stock = []
        stock = yf.download(i,start=start, end=end, progress=False)
        
        # append the individual stock prices 
        if len(stock) == 0:
            None
        else:
            stock['Name']=i
            stock_final = stock_final.append(stock,sort=False)
    except Exception:
        None
```

```python
stock_final.head()
```

```python
import pandas as pd
symbols_to_save = ['AMZN', 'FB', 'GOOG', 'AMD']
symbol_dataframes = {}

for symbol in symbols_to_save:
    symbol_df = stock_final[stock_final['Name'] == symbol]
    symbol_dataframes[symbol] = symbol_df

amzn_df = symbol_dataframes['AMZN']
fb_df = symbol_dataframes['FB']
goog_df = symbol_dataframes['GOOG']
amd_df = symbol_dataframes['AMD']

amzn_df.to_csv('AMZN_stock_data.csv', index=False)
fb_df.to_csv('FB_stock_data.csv', index=False)
goog_df.to_csv('GOOG_stock_data.csv', index=False)
amd_df.to_csv('AMD_stock_data.csv', index=False)
```
